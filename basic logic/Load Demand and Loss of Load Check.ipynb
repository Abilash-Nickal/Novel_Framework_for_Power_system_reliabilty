```python
# -----------------------------
# Simple Base Load Dispatch
# -----------------------------

# Generator Data
gen_capacity = [300, 200, 150]      # MW
is_coal      = [False, True, False]  # Coal or not
is_up        = [True, True, True]   # Generator availability
unit_cost    = [20, 30, 25]         # LKR per kWh

current_load = 500   # MW

dispatched_power = 0
hourly_cost = 0

# Loop through generators
for i in range(len(gen_capacity)):

    # Skip if generator is not coal
    if not is_coal[i]:
        continue
    print("Generator", i, "is not coal, skipping.")

    # Skip if generator is down
    if not is_up[i]:
        continue
    print("Generator", i, "is down, skipping.")

    # Stop if load already satisfied
    if dispatched_power >= current_load:
        break

    # Calculate remaining power needed
    needed = current_load - dispatched_power

    # Generator contribution
    contribution = min(gen_capacity[i], needed)

    # Add power
    dispatched_power += contribution

    # Calculate cost
    hourly_cost += contribution * 1000 * unit_cost[i]

    print("Generator", i, "supplied", contribution, "MW")

print("Total Dispatched:", dispatched_power, "MW")
print("Total Cost:", hourly_cost, "LKR")
```
