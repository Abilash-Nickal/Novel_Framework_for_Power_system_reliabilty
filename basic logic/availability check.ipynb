```python
import random
import csv # Import the CSV library
#print import os   Used for finding the home directory
# Generator capacities and forced outage rates
Gen = [10, 20,30,40,50,60]
FOR = [0.01, 0.02,0.01, 0.02, 0.03, 0.04]
```

```python
availableGen = 0
if random.random() > FOR[0]:
        availableGen += Gen[0]
if random.random() > FOR[1]:
        availableGen += Gen[1]
if random.random() > FOR[2]:
        availableGen += Gen[2]
if random.random() > FOR[3]:
        availableGen += Gen[3]
if random.random() > FOR[4]:
        availableGen += Gen[4]
if random.random() > FOR[5]:
        availableGen += Gen[5]
print(availableGen)
```

```python
# List to store the available capacity from each of the 10 checks
results = []
```

```python
for check_num in range(10000000):
    availableGen = 0

    # Check each generator (using indices 0 through 5)
    # Note: This is a series of independent checks based on the indices
    # you provided in your original code.

    # Check for FOR[0] and add Gen[0] (10 MW)
    if random.random() > FOR[0]:
        availableGen += Gen[0]

    # Check for FOR[1] and add Gen[1] (20 MW)
    if random.random() > FOR[1]:
        availableGen += Gen[1]

    # Check for FOR[2] and add Gen[0] (10 MW)
    if random.random() > FOR[2]:
        availableGen += Gen[2]

    # Check for FOR[3] and add Gen[1] (20 MW)
    if random.random() > FOR[3]:
        availableGen += Gen[3]

    # Check for FOR[4] and add Gen[0] (10 MW)
    if random.random() > FOR[4]:
        availableGen += Gen[4]

    # Check for FOR[5] and add Gen[1] (20 MW)
    if random.random() > FOR[5]:
        availableGen += Gen[5]
```

```python
# Loop 10 times to perform 10 separate system availability checks

    # Store the result of this check
results.append(availableGen)

print("Available capacity for 10 consecutive checks:")
print(results)
```

The statement results.append(availableGen) is a fundamental Python operation that means: Take the current value stored in the variable availableGen and add it to the end of the list named results.

```python
if results:
    # 1. Sum all values in the list
    total_capacity_sum = sum(results)

    # 2. Divide by the number of checks (the length of the list)
    average_available_gen = total_capacity_sum / len(results)

    print(f"Total number of checks: {len(results):,}")
    print(f"Total capacity sum across all checks: {total_capacity_sum:,.2f} MW")
    print(f"\nAverage Available Generation: {average_available_gen:.2f} MW")
else:
    print("No simulation data available to calculate the average.")
```
